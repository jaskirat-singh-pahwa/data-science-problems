```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
## GENERATING LABEL AND TARGETS
## ITS A SIMPLE LINEAR REGRESSION, SO THERE IS ONLY ONE FEATURE VARIABLE AND ONE TARGET VARIABLE

## numpy.arange() returns evenly spaced values within a given interval.
## numpy.arange(start , stop , step , dtype = None) {It Returns numpy ndarray}.

X = np.arange(1 , 11).reshape(10,1)
y = np.array([7, 8, 7, 13, 16, 15, 19, 23, 18, 21]).reshape(10,1)
```

```python
## PLOTTING THE LABELS AND TARGETS

plt.plot(X , y ,'ro')
plt.xlabel('LABELS')
plt.ylabel('TARGETS')
plt.title('LINEAR REGRESSION')
plt.grid()
plt.show()
```

```python
## INITIALIZING LINEAR REGRESSION
model = LinearRegression()

## TRAINING THE MODEL
model.fit(X,y)

print("Model Coefficient: " ,model.coef_)
print("Model Intercept: " ,model.intercept_)
```

```python
## LETS COMPUTE PREDICTION VECTOR A USING COEFFICIENT AND INTERCEPT
a = model.coef_ * X + model.intercept_       ## a is our linear regression model function

## LETS PLOT IT ON GRAPH
plt.plot(X , y ,'ro' , X , a)
axes = plt.gca()      ## Gets the current axes
axes.set_ylim([0 , 25])
plt.xlabel('LABELS')
plt.ylabel('TARGETS')
plt.title('LINEAR REGRESSION')
plt.grid()
plt.show()
```

```python
## ACCURACY OVER TRAINING DATA
print(model.score(X,y))
```

```python
### FEATURES ARE SOMETHING WE USE TO DINTINGUISH ONE OBJECT FROM OTHER. SO OBVIOUSLY, IF WE HAVE MORE FEATURES, 
### THEN WE WILL LIKELY TO HAVE A BETTER FIT MODEL
### BUT HOW WE CAN ACQUIRE MORE FEATURES?

### THE EASIEST WAY IS TO ADD POLYYNOMIAL FEATURES:
### THE EASIEST WAY TO ADD MORE FEATURES, IS TO COMPUTING POLYNOMIAL FEATURES FROM THE PROVIDED FEATURES.
### IT MEANS THAT IF WE HAVE X THEN WE CAN USE Xsquare OR Xcube etc.

X = np.c_[X , X**2]    ## np.c_ : translates slice objectsto concatenation along second axis

X
```

```python
model.fit(X , y)

x = np.arange(1 , 11 , 0.1)
x = np.c_[x , x**2]
a = np.dot(x, model.coef_.transpose()) + model.intercept_
```

```python
# x = x.reshape(20,2)
#y = y.reshape(1,20)
plt.plot(X[:, 0] , y , 'ro' , x[:, 0] , a)
plt.show()
```

```python
model.score(X,y)## Returns the coefficient of determination R^2 of prediction, means the goodness of fit.
```
